```python
df = pd.DataFrame([])
i = 0
while True:
    start = i*feature_num
    end = (i+1)*feature_num
    if end > len(df_train): break
#     if i == 3: break
    
    date = df_train[start:end].iloc[0,0]
    datetime_list = pd.date_range(start=date, periods=24, freq='60min')
    df_dummy = df_train[start:end].T[3:]
    df_dummy.columns = feature_list
    df_dummy['time'] = datetime_list
#     display(df_dummy)
    df = df.append(df_dummy)
    
    i+=1

df
```

```python
col_list = list(df.columns)
col_list = col_list[-1:] + col_list[:-1]
df = df[col_list]
df
```

```python
# turn NR to 0
def return_val(x):
    if x == 'NR': return 0
    else: return float(x)

for col in df.columns:
    if col !='time':
        df[col] = df[col].apply(lambda x: return_float(x))

df
```

```python
df = df.reset_index(drop=True)
df
```

# Slice one month then slice group by group

```python
day_per_month = 20
hr_per_day = 24
hr_per_month = day_per_month * hr_per_day
hr_per_group = 10

x_train = []
y_train = []

for i_month in range(0, len(df), hr_per_month):
    
    start_month = i_month
    end_month = i_month + hr_per_month
    print(start_month, end_month)
    df_one_month = df[start_month:end_month]
    
#     display(df_one_month)
    
    start_group = 0
    while True:
        
        end_group = start_group + hr_per_group
        if end_group > len(df_one_month): break
        
        df_one_group = df_one_month.iloc[start_group:end_group, :]
#         display(df_one_group)
        start_group+=1
    
        x = df_one_group.iloc[:-1, 1:].values # not include the 10th hour's data
        y = df_one_group['PM2.5'].values[-1] # 10th hour's PM2.5
        ### DataFrame.to_numpy() == DataFrame.values
        
#         display(x)
        print('PM2.5 of {}th hour:'.format(i_month+end_group), y)
        
        x = x.T
#         display(x)
        
        x_train.append(x)
        y_train.append(y)
        
x_train = np.array(x_train)
y_train = np.array(y_train)

x_train.shape, y_train.shape # '(groups, rows_every_group, columns_every_group), (columns,)'
```

```python
x_train = x_train.reshape(len(x_train),-1) # reshape to a 1D list
print(x_train.shape)
bias = np.ones(shape=(len(x_train),1))
print(x_train.shape)
x_train = np.concatenate([x_train,bias],axis=1)
print(x_train.shape)
```

# Training

```python
learning_rate = 10
epochs = 100
gradient_square_sum = np.zeros(shape=x_train.shape[1])
w = np.zeros(shape=x_train.shape[1])
```

```python
for epoch in range(epochs):
    y_hat = x_train @ w
    delta_y = y_hat - y_train
#     print('delta_y:', delta_y)
    gradient = x_train.T @ delta_y
    
    gradient_square_sum = gradient_square_sum + gradient ** 2
    ada = np.sqrt(gradient_square_sum) ### adagrad
    new_gradient = gradient / ada
    w = w - learning_rate * new_gradient
    
    mse = (np.sum(delta_y**2))/len(delta_y) ### mean square error
    rmse = np.sqrt(mse)
    print('==============================================')
    print(epoch, rmse, mse)
```
