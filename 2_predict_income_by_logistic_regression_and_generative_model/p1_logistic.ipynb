## Normalization

https://morvanzhou.github.io/tutorials/machine-learning/sklearn/3-1-normalization/

```python
def normalize(df_array):
    mean = df_array.mean(axis=0)
    std = df_array.std(axis=0)
    
    # To prevent from divide by inf: if std = 0, replace it with 1
    std[std == 0] = 1
    
    df_array_normalized = (df_array - mean) / std
    return df_array_normalized
```

# Training

```python
learning_rate = 0.1
```

```python
w = np.zeros(shape=x_train_bias.shape[1])
w.shape
```

```python
z_train = x_train_bias @ w
z_train.shape, x_train_bias.shape, w.shape
```

```python
def sigmoid(z):
    res = 1 / (1.0 + np.exp(-z))
    return np.clip(res, 1e-8, 1-(1e-8))

y_hat = sigmoid(z_train)
y_hat.shape, z_train.shape
```

```python
y_delta = y_train - y_hat
print(y_delta.shape, y_train.shape, y_hat.shape)
print(y_delta[0], y_train[0], y_hat[0])
```

```python
gradient = x_train_bias.T @ y_delta
gradient.shape, x_train_bias.T.shape, y_delta.shape
```

```python
w = w - learning_rate * gradient
w.shape
```

```python
epochs = 100
batch_size = 32
learning_rate = 0.0001
# print(x_train_bias.shape[1])
w = np.zeros(shape=x_train_bias.shape[1])

for epoch in range(epochs):
    
    x_train_bias, y_train = shuffle(x_train_bias, y_train)
    
    batch_num = int(len(x_train_bias) / batch_size)
    
    # update parameters
    z_train = x_train_bias @ w
    y_hat = sigmoid(z_train)
    y_delta = y_train - y_hat
    gradient = - x_train_bias.T @ y_delta
    w = w - learning_rate * gradient
    
    # calculate loss 
    y_real_0 = y_train
    y_real_1 = 1 - y_train
    y_predict_0 = y_hat
    y_predict_1 = 1 - y_hat
    cross_entropy = - y_real_0 * np.log(y_predict_0) - y_real_1 * np.log(y_predict_1)
    total_cross_entropy = np.sum(cross_entropy)
    mean_cross_entropy = np.mean(cross_entropy)
    print(total_cross_entropy, mean_cross_entropy)
```

```python
epochs = 100
batch_size = 32
learning_rate = 0.0001

# split (train, val)
split_percent = 0.1
num_sample = len(y_train)
num_sample_train = int(num_sample * (1 - split_percent))
x_train_split = x_train_bias[:num_sample_train]
y_train_split = y_train[:num_sample_train]
x_val_split = x_train_bias[num_sample_train:]
y_val_split = y_train[num_sample_train:]

w = np.zeros(shape=x_train_split.shape[1])
batch_num = len(y_train_split) // batch_size

for epoch in range(epochs):
    
    x_train_split, y_train_split = shuffle(x_train_split, y_train_split)
    
    total_cross_entropy = 0
    total_sample = 0
    
    for i_batch in range(batch_num):
        
        i_batch_start = i_batch * batch_size
        i_batch_end = (i_batch + 1) * batch_size
        
        x_batch = x_train_split[i_batch_start:i_batch_end]
        y_batch = y_train_split[i_batch_start:i_batch_end]
        
        # update parameters
        z_train = x_batch @ w
        y_hat = sigmoid(z_train)
        y_delta = y_batch - y_hat
        gradient = - x_batch.T @ y_delta
        w = w - learning_rate * gradient
        
        # calculate loss
        y_real_0 = y_batch
        y_real_1 = 1 - y_batch
        y_predict_0 = y_hat
        y_predict_1 = 1 - y_hat
        cross_entropy = - y_real_0 * np.log(y_predict_0) - y_real_1 * np.log(y_predict_1)
        
        total_cross_entropy += np.sum(cross_entropy)
        total_sample += len(y_batch)
        
    # calculate accuracy
    z_val = x_val_split @ w
    y_val_predict = sigmoid(z_val)
    y_val_predict_category = np.round(y_val_predict)
    compare = (y_val_predict_category == y_val_split)
    accuracy = np.sum(compare) / len(x_val_split)
    
    # mean cross entropy
    mean_cross_entropy = total_cross_entropy / total_sample
    
    # training log
    log = np.array([total_cross_entropy, mean_cross_entropy, accuracy, total_sample])
    log = ['%.3f'%(x) for x in log]
    print(log)
```
